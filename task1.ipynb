```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv(r"C:\Users\padmamahe\OneDrive\Desktop\data.csv")

# Display first 5 rows
print("First 5 rows:")
print(df.head())

# Data Info
print("\nDataset Info:")
print(df.info())

# Describe data
print("\nStatistical Summary:")
print(df.describe())

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Check data types
print("\nData Types:")
print(df.dtypes)

# Shape of dataset
print("\nShape of dataset:")
print(df.shape)

# Unique values in categorical columns
print("\nUnique values:")
for col in df.select_dtypes(include='object'):
    print(f"{col}: {df[col].unique()}")
```
